```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Task 1: Load the dataset using Pandas and display the first few rows to understand its structure.
data = pd.read_csv(r'C:\Users\TB352CU\Downloads\sales_data.csv')
data
```

```python
# Task 2: Perform data cleaning: • Check for missing values and decide on a strategy to handle them. • Identify and remove any duplicate rows.
```

```python
data.head()
```

```python
data.info()
```

```python
data.nunique()
```

```python
data.drop_duplicates(inplace = True)
```

```python
# Task 3: Compute and display the following summary statistics: • Mean, median, and standard deviation of the "Quantity" and "Price" columns. • Total sales revenue (computed as the sum of Quantity * Price).
```

```python
data[['Quantity','Price']].describe()
```

```python
data['Revenue'] = data['Quantity']*data['Price']
data
```

```python
data['Revenue'].sum()
```

```python
# Task 4: Visualize the data: • Create a histogram of the "Quantity" column to understand the distribution of sales quantities. • Generate a scatter plot of "Price" vs. "Quantity" to explore any potential relationships.
```

```python
fig,ax = plt.subplots(1,2, figsize = (15,5))
ax[0].hist(data['Quantity'])
ax[0].set_xlabel('Quantity')

ax[1].scatter(data['Price'],data['Quantity'])
ax[1].set_xlabel('Price')
ax[1].set_ylabel('Quantity')

plt.plot()
```

```python
# Task 5: Perform categorical analysis: • Group the data by "Product" and compute the total quantity sold and average price for each product. • Visualize the product-wise sales using a bar chart.
```

```python
df = data.groupby('Product')[['Quantity','Price']].agg(['sum','mean'])
df
```

```python
product_sales = df['Quantity']['sum']
product_sales.plot(kind='bar')
```

```python
# Task 6: Identify outliers: • Use the z-score method to detect potential outliers in the "Price" column. • Visualize the outliers using a box plot.
```

```python
data['z_scores'] = (data['Price'] - data['Price'].mean())/ data['Price'].std()
data
```

```python
data[data['z_scores'] > 3] 
```

```python
data['z_scores'] < -3
```

```python
import seaborn as sns
sns.boxplot(x=data['Price'] )
```

```python
# Task 7: Additional Analysis: • Calculate the weekly sales by aggregating the data based on the "Date" column. • Visualize the weekly sales trend over time using a line plot.
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
weekly_sales = data.resample('W', on='Order Date').sum()
# Separate x and y values
x_values = weekly_sales.index
y_values = weekly_sales.values

plt.plot(x_values, y_values)
plt.title("Weekly Sales Trend")
plt.xlabel("Week")
plt.ylabel("Total Sales")
plt.show()
```
